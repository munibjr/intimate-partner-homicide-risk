# 01: Problem Context & Data Exploration

## The Norwegian Problem

**71% of intimate partner homicides had prior registered partner violence**

Yet cases are often treated as isolated incidents by separate agencies. This notebook explores the dataset and visualizes the escalation patterns that go unrecognized.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.case_generator import CaseGenerator
from src.data.feature_extractor import FeatureExtractor

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

### Generate Dataset

```python
# Generate synthetic dataset
gen = CaseGenerator(seed=42)
cases = gen.generate_dataset(n_cases=500)

# Extract features
extractor = FeatureExtractor()
features_list = [extractor.extract_features(c) for c in cases]

print(f"Total cases: {len(cases)}")
print(f"High-risk: {sum(1 for c in cases if c['risk_category'] == 'high')}")
print(f"Low-risk: {sum(1 for c in cases if c['risk_category'] == 'low')}")
```

### Risk Distribution

```python
# Create risk distribution visualization
risk_labels = np.array([f['risk_label'] for f in features_list])

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Risk label distribution
ax1.hist(risk_labels, bins=20, alpha=0.7, color='darkred', edgecolor='black')
ax1.set_xlabel('Risk Score (0-1)')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of Risk Scores')
ax1.axvline(0.4, color='orange', linestyle='--', label='Low/Medium threshold')
ax1.axvline(0.7, color='red', linestyle='--', label='Medium/High threshold')
ax1.legend()

# Risk category breakdown
categories = [c['risk_category'] for c in cases]
category_counts = pd.Series(categories).value_counts()
colors = {'high': 'darkred', 'low': 'darkgreen'}
ax2.bar(category_counts.index, category_counts.values, color=[colors.get(cat, 'gray') for cat in category_counts.index])
ax2.set_ylabel('Count')
ax2.set_title('Risk Category Distribution')
for i, v in enumerate(category_counts.values):
    ax2.text(i, v + 5, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
